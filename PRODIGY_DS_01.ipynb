# Categorical Variables Distribution
# Histogram

#The categorical data in this dataset include of - Gender, Marital Status, Occupation, Monthly Income, Educational Qualifications, Output, Feedback and Online Food Consumption.

#For Histogram we will be using Gender, Marital Status, Occupation and Monthly Income

```python
#1. Gender in the Online Food Consumption Population

plt.figure(figsize=(5,3))
plt.hist(data['Gender'], bins=3)
plt.title("Genders of people ordering food online") 
plt.xlabel("Categories of Genders") 
plt.ylabel("Number of Genders")
plt.show
```

```python
#2. Marital Status in the Online Food Consumption Population

plt.figure(figsize=(5,3))
plt.hist(data['Marital Status'], bins=5, color='red', alpha=0.45)
plt.title("Marital Status of people ordering food online") 
plt.xlabel("Categories of Marital Status") 
plt.ylabel("Number of Marital Status")
plt.show
```

```python
#3. Occupation in the Online Food Consumption Population

plt.figure(figsize=(5,3))
plt.hist(data['Occupation'], bins=7, color='green', alpha=0.45)
plt.title("Occupation of people ordering food online") 
plt.xlabel("Categories of Occupation") 
plt.ylabel("Number of Occupation")
plt.show
```

```python
#4. Monthly Income in the Online Food Consumption Population

plt.figure(figsize=(9,3))
plt.hist(data['Monthly Income'], bins=9, alpha=0.45)
plt.title("Monthly Income of people ordering food online") 
plt.xlabel("Categories of Monthly Income") 
plt.ylabel("Counts of Monthly Income")
plt.show
```

```python
#Using all 4 of them together - Consider zooming in for better readability

plt.figure(figsize=(25,5))
plt.hist(data['Gender'],bins = 3, alpha = 0.5,  
         color = 'red') 
plt.hist(data['Marital Status'],bins = 5, alpha = 0.5, 
         color = 'blue') 
plt.hist(data['Occupation'],bins = 7, alpha = 0.5, 
         color = 'yellow') 
plt.hist(data['Monthly Income'],bins = 9, alpha = 0.5, 
         color = 'green') 
  
plt.title("histogram with multiple variables") 
plt.legend(['Gender','Marital Status','Occupation','Monthly Income']) 
plt.xlabel("Categories of Data") 
plt.ylabel("Values of Data")
  
plt.show() 
```

# Categorical Variables Distribution
# Using different Bar Values

#The categorical data in this dataset include of - Gender, Marital Status, Occupation, Monthly Income, Educational Qualifications, Output, Feedback and Online Food Consumption.

#For Bar Graph we will be using Educational Qualifications, Output, Feedback and Online Food Consumption individually or in combinations.

```python
#1. Educational Qualifications by Ages of people ordering food online

plt.figure(figsize=(9,5))
plt.bar(data['Age'], data['Educational Qualifications'], color='g')
plt.title("Educational Qualifications by Ages of people ordering food online") 
plt.xlabel("Ages of Individuals") 
plt.ylabel("Types of Educational Qualifications")
plt.show
```

# Continuous Variables Distribution
# Histogram

#The continuous data in this dataset include of - Age, Family size, latitude, longitude, and Pin code.

#For Histogram we will be using Age,and Family size.

```python
#1. Age in the Online Food Consumption Population

plt.figure(figsize=(9,3))
plt.hist(data['Age'], bins=31, alpha=0.8, color='yellow')
plt.title("Age of people ordering food online") 
plt.xlabel("various Ages") 
plt.ylabel("Count of people")
plt.show
```
